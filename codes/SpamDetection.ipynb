```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB,BernoulliNB
```

```python
data= pd.read_csv("/content/drive/Othercomputers/My Computer/Machine Learning/Datasets/sms_spam.csv")
```

```python
data.head(10)
```

```python
data.groupby('type').describe()
```

```python
data.columns
```

```python
data['spam']=data['type'].apply(lambda x:1 if x=="spam" else 0 )
```

```python
data.head()
```

```python
X_train,X_test,y_train,y_test=train_test_split(data.text,data.spam,random_state=2)
```

```python
cv=CountVectorizer(binary=True)
X_train_Matrix=cv.fit_transform(X_train.values)
X_train_Matrix.shape
```

```python
model1=BernoulliNB()

model1.fit(X_train_Matrix,y_train)

```

```python
X_test_Matrix=cv.transform(X_test.values)
```

```python
model1.score(X_test_Matrix,y_test)
```

```python
new_message=["This weekend i am busy","you won lucky draw"]
new_message_matrix=cv.transform(new_message)
```
